#Cropping the image in the next line of code

Let's do cropping to bigger photo,because photograph is too big.

```
image2 = Image.open('mine.jpg')
box = (200,350,900,1000)
cropped_image = image2.crop(box)
cropped_image.save('cropped_image.jpg')
print(cropped_image.size)
```

```
cropped_image
```

```
cropped_image = Image.open('cropped_image.jpg')
red,blue,green = cropped_image.split()

print(cropped_image.mode) # Output: RGB
print(red.mode) # Output: L
print(blue.mode) # Output: L
print(green.mode)# Output: L

new_image = Image.merge("RGB",(red,green,blue))
new_image.save('new_image.jpg')

print(new_image.mode)
```

```
new_image
```

#Enhancing the photograph

```
image3 = Image.open('image_400.jpg')

contrast = ImageEnhance.Contrast(image3)
contrast.enhance(1.5).save('contrast.jpg')
```

```
new=cv2.imread('contrast.jpg',1)
cv2_imshow(new)
cv2.waitKey(0)
```

#Smoothing the image

```
gray_image = cv2.cvtColor(new,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
```

```
inverted_image=255-gray_image
cv2_imshow(inverted_image)
cv.waitKey(2)
```

Now the next step in the process is to blur the image by using the Gaussian Function in OpenCV:

```
blurring=cv2.GaussianBlur(inverted_image,(21,21),5)
inverted_Blurred=255-blurring
pencil_sketch=cv2.divide(gray_image,inverted_image,scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
